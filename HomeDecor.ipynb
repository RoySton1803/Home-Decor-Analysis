```python
import pandas as pd
import numpy as np
```

```python
#loading data 
df = pd.read_csv("india_home_decor_uncleaned.csv")
df.head()
```

```python
#handling missing values by using median and filling values 
df["Age Group"].fillna("Unknown", inplace=True)
df["Preferred Style"].fillna("Unknown", inplace=True)
df["Purchase Frequency"].fillna(df["Purchase Frequency"].median(), inplace=True)
df["Average Spend (INR)"].fillna(df["Average Spend (INR)"].median(), inplace=True)
df["Top Competitor Chosen"].fillna("Unknown", inplace=True)
df["Influencing Factor"].fillna("Unknown", inplace=True)
df["Purchase Channel"].fillna("Unknown", inplace=True)
df["City"].fillna("Unknown", inplace=True)
```

```python
df.head(10)
```

```python
style_corrections = {"minimalst": "Minimalist", "bohemn": "Bohemian"}
df["Preferred Style"] = df["Preferred Style"].replace(style_corrections)
```

```python
competitor_corrections = {
    "ikea ind": "IKEA India",
    "pepper fry": "Pepperfry",
    "urban ladder": "Urban Ladder",
    "home town": "HomeTown"
}

factor_corrections = {
    "pricee": "Price",
    "Quality ": "Quality",
    "brand reputaion": "Brand Reputation",
    "social influenc": "Social Influence"
}

channel_corrections = {
    "in store": "In-Store",
    "whatsapp": "WhatsApp",
    "insta": "Instagram"
}

city_corrections = {
    "Banglore": "Bangalore",
    "mumbai ": "Mumbai",
    "del hi": "Delhi"
}

# Apply corrections
df["Top Competitor Chosen"] = df["Top Competitor Chosen"].replace(competitor_corrections)
df["Influencing Factor"] = df["Influencing Factor"].replace(factor_corrections)
df["Purchase Channel"] = df["Purchase Channel"].replace(channel_corrections)
df["City"] = df["City"].replace(city_corrections)
```

```python
df.head(10)
```

```python
df.drop_duplicates(inplace=True)
```

```python
#integer and round off 
df["Average Spend (INR)"] = df["Average Spend (INR)"].astype(float)
df["Purchase Frequency"] = df["Purchase Frequency"].astype(int)
```

```python
df.head(10)
```

```python
# Save cleaned dataset (optional)
df.to_csv("home_decor_cleaned.csv", index=False)
```
